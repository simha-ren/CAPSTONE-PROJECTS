```python
from transformers import pipeline
```

```python
# Initialize a sentiment-analysis pipeline
classifier=pipeline("sentiment-analysis")
sentences=[
    "The client was very satisfied with the delivery. ",
    "The project is significantly over budget and behind schedule. ",
    "The new regulatory framework presents both risks and opportunities.",
    "The stakeholder8 feedback indicates strong alignment with the proposed strategy."
]
```

```python
for sentence in sentences:
    result=classifier(sentence)
    label=result[0]['label']
    score=result[0]['score']
    print(f'"Sentence: {sentence[:50]}..."--> {label}({score:.4f})')
```

```python
def get_sentiment(classifier, sentence):
    result=classifier(sentence)
    return result[0]['label'], result[0]['score']
```
